```python
# set index by Date

df = pd.read_csv(r"C:\Users\ASYIKIN\OneDrive\Desktop\MySQL - DA\Panda - Project 6\datasets\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

# PLOT

```python
# SET has to come BEFORE creating the plot

print(plt.style.available)
plt.style.use('seaborn-v0_8-dark-palette')

df.plot(kind = 'line', 
        title = 'Ice Cream Rating', 
        xlabel= 'Daily Ratings', 
        ylabel = 'Scores')
```
