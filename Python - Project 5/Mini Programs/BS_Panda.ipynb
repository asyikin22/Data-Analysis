```python
from bs4 import BeautifulSoup
import requests

url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_Malaysia'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')

table = soup.find_all('table')[1]
```

```python
titles = table.find_all('th')

#list comprehension to store the table names
column_headers = [title.text.strip() for title in titles]
print('Column Headers:')
print(f'{column_headers}\n')

for header in column_headers:
    print(f'- {header}')
```

```python
import pandas as pd

df = pd.DataFrame(columns = column_headers)
df
```

```python
column_data = table.find_all('tr')

for row in column_data:
    row_data = row.find_all('td')
    each_row_data = [data.text.strip() for data in row_data]
    
     #this will get rid of empty list
    if each_row_data:                                  
        print(each_row_data)
      
```

```python
import pandas as pd

df = pd.DataFrame(columns = column_headers)

column_data = table.find_all('tr')

for row in column_data:
    row_data = row.find_all('td')
    each_row_data = [data.text.strip() for data in row_data]
    
     #this will get rid of empty list
    if each_row_data:                                  
        df.loc[len(df)] = each_row_data

df 
```
